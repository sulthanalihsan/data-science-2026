```python
# Operasi aritmetika vektorisasi
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])
print(a + b) # → [11 22 33 44 55]
print(a * b) # → [10 40 90 160 250]
print(a ** 2) # → [1 4 9 16 25]
# Universal functions (ufuncs)
np.sqrt(a) # akar kuadrat tiap elemen
np.exp(a) # eksponensial (e^x)
np.log(b) # logaritma natural
# Fungsi statistik
data = rng.normal(0, 1, 1000)
# print(data)
print(f'Mean: {np.mean(data):.4f}')
print(f'Std: {np.std(data):.4f}')
print(f'Min: {np.min(data):.4f}')
print(f'Max: {np.max(data):.4f}')
print(f'Q1: {np.percentile(data,25):.4f}')
print(f'Q3: {np.percentile(data,75):.4f}')
```

```python
# Menggunakan link raw agar bisa dibaca langsung oleh pandas
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# 2. Eksplorasi dasar
print(df.shape) # (891, 12)
print(df.head()) # 5 baris pertama
print(df.tail(3)) # 3 baris terakhir
print(df.info()) # tipe & missing
print(df.describe()) # statistik deskriptif
# 3. Cek missing values
print(df.isnull().sum()) # per kolom
print(df.isnull().mean()) # proporsi %

# Menampilkan 5 data teratas untuk memastikan data termuat
df.head()
```

```python
# .loc — label-based
df.loc[0, 'Name'] # row 0, kol Name
df.loc[0:4, 'Age':'Fare']
# .iloc — integer-based
df.iloc[0, 2] # baris 0, kolom 2
df.iloc[0:5, 0:3] # 5 baris, 3 kolom
# Modifikasi via .loc
df.loc[df['Age']<18, 'Kategori']= 'Anak-anak'
```

```python
# query() — sintaksis SQL-like
df.query('Age > 18 and Pclass==1')
# # isin() — filter multi-nilai
# kota = ['Jakarta','Bandung']
# df[df['Kota'].isin(kota)]
# # str accessor
# df[df['Name'].str.contains('Miss')]
# df['Name'].str.lower()
```

```python
# GroupBy dasar
# Rata-rata usia per kelas penumpang
df.groupby('Pclass')['Age'].mean()
# Beberapa agregasi sekaligus
df.groupby('Pclass').agg(
jumlah=('Survived','count'),
selamat=('Survived','sum'),
rata_usia=('Age','mean')
)
# GroupBy multi-kolom
df.groupby(['Pclass','Sex'])['Survived'].mean().unstack()
# Pivot Table (seperti Excel pivot)
pd.pivot_table(df,
values='Survived',
index='Pclass', columns='Sex',
aggfunc='mean')
```
